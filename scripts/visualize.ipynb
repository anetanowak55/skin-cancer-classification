```python
def addlabels(ax):
  for rect in ax.patches:
      height = rect.get_height()
      ax.text(
          rect.get_x() + rect.get_width() / 2, height + 5, rect.get_height(), ha="center", va="bottom", Bbox = dict(facecolor = 'grey', alpha =.6)
      )
```
